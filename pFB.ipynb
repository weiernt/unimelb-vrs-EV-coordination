# The Preconditioned Forward-Backward Algorithm (Alg. 1)
Initial imports

```python
import numpy as np
import cvxopt
import matplotlib.pyplot as plt
import time
rng = np.random.default_rng()
```

```python
%matplotlib notebook
def get_cmap(n, name='gist_rainbow'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)

# model the non-PEV demand
def non_pev_demand(t):
    return -1.5*np.sin((1/12)*np.pi*t + 2.5*np.pi*(1/12)) + 7.5

def init_xbar(size : int):
    xbar = np.zeros((size,1))
    for i in range(size):
        num = rng.random()
        if (num<=0.8):
            xbar[i] = rng.uniform(low=1, high=5)
    return xbar

class Agent:
    def __init__(self, m):
        """constructor, m is the number of rows in vector x, or the time period for this experiment (ie: 24 for 24hrs)"""
        self.Q_i     = np.diag(rng.uniform(low=0.1, high=4, size=m))
        self.p_i     = np.vstack(rng.uniform(low=0.2, high=2, size=m)) # already transposed, ie: p_i is a vertical vect
        self.x_bar_i = init_xbar(m)
        self.l_i     = rng.uniform(low=0.5, high=1.5)

        # self.lambd_k = rng.random((m,1))
        self.x_k     = rng.random((m,1))  # represents x_k^0, only inits them to be from [0,1)
        # self.alpha_i = 1/(1 + delta)

    def factory_create(n, m):
        """ create and randomly initialise appropriately, 'n' amounts of agents, each agent has "m" time horizons, and delta,
        returns them as an array"""
        return [Agent(m) for i in range(n)]

    def cost_function(self, avg_x, c):
        return 0.5 * (self.x_k.T @ self.Q_i @ self.x_k) + (self.p_i.T @ self.x_k) + ((avg_x + c).T@self.x_k) 
    
    def analyse_self(self):
        print(f"sum of x_k = {sum(self.x_k)}")
        print(f"l_i = {self.l_i}")

if __name__ == "__main__":
    # plt.figure(figsize=((6.4 * 1.6,4.8 * 1.6)))
    plt.figure()
    plt.plot(np.arange(0,24), non_pev_demand(np.arange(0,24)))
#     plt.title("Non PEV demand over a 24 hour period")
#     plt.xlabel("time (hour)")
#     plt.ylabel("Power (kW)")
    plt.grid()
#     plt.savefig("non_pev_demand.png")
```

```python
m = 24
N = 20
EPSILON = 1e-5

def prox(agent, alpha_i, y):
    """The Proximal operator, needs an Agent object, and the function being acted on, 
    will return the 'x' that achieves the minimum"""
    # Gx <= h
    m = len(agent.x_k)
    G = np.block([[np.identity(m)], 
                [-np.identity(m)], 
                [-np.ones((1,m))]])

    h = np.block([[agent.x_bar_i],
                [np.zeros((m,1))],
                [-agent.l_i]])

    # transform the np.arrays into cvxopt.matrix
    P = cvxopt.matrix(agent.Q_i + (1/alpha_i) * np.identity(np.shape(agent.Q_i)[0]))
    q = cvxopt.matrix(agent.p_i - (1/alpha_i) * y)
    h = cvxopt.matrix(h)
    G = cvxopt.matrix(G)
    
    output = cvxopt.solvers.coneqp(P, q, G, h, options={"show_progress" :False})["x"]

    return np.array(output)

def plot_agents(list_of_agents, save, legend=False, n=None):
    plt.figure(figsize=((6.4 * 1.4,4.8 * 1.4)))
    ax = plt.subplot(111)
    prev_y = np.zeros(m)
    i=0

    cmap = get_cmap(len(list_of_agents))
    for agent in list_of_agents:
        ax.bar(np.arange(0,m), unraveled_xk := np.ravel(np.array(agent.x_k)), width=1, bottom=prev_y,
                label=f"agent[{i}]",
                color=cmap(i),
                edgecolor="black")
        prev_y += unraveled_xk
        i+=1
        
    if (n==None):
        plt.title(title:=f"Final Plot of all (N={N}) decision variables $x_i$,\n with respect to the time period, using pFB")
    else:
        plt.title(title:=f"Current iteration = {n},\n Plot of all (N={N}) decision variables $x_i$,\n with respect to the time period, using pFB")
    plt.xlabel("time horizon (hour)")
    plt.ylabel("Power (kW)")
    
    # Shrink current axis by 20%
    box = ax.get_position()

    # Put a legend to the right of the current axis
    if legend:
        ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
        ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
    
    if save==True:
        plt.savefig(f"./images/{N}_agents_iteration_{n}.png")
```

```python
list_of_agents = Agent.factory_create(N, m=m)

# perform the pFB
def perform_pfb(show_progress):
    # m = number of rows in vector x, time horizon
    # n = number of rows in matrix A, in this case, it is equal to m
    # N = number of agents

    ZEROS_VECTOR = np.zeros((m,1))

    # A = np.kron(np.ones((1,N)), np.eye(m))  # big A that represents the coupling constraints on agents
    # then norm of A_i  (||A_i||)= 1
    A_i = np.eye(m)

    # cocoercive constant
    gamma = N/((N**2) + 2*N + 1)

    #select delta, since delta > 1/2*gamma
    delta = 1/(2*gamma)

    # just initialise alpha globally since they are all the same
    alpha_i = (1/(1 + delta)) * 0.99

    # the constant c, which appears in the function f_i, samples the d(t) function at discrete intervals
    c = np.vstack([non_pev_demand(i) for i in range(m)])

    # affine coupling constraint. (used with A, in Ax-b<=0)
    b_i = np.full((m,1), fill_value=0.55) # using 0.55kW   (from the function K(t))

    # init beta
    beta = (1/(1 + (delta/N))) * 0.99

    # theoretical Lipschitz and strong monotonicity constant:
    # lip = 1 + (1/N)
    # strong_mono = 1/N

    # initialise the agents
#     list_of_agents = Agent.factory_create(N, m=m)

    avg_x = (1/N)*sum([list_of_agents[i].x_k for i in range(len(list_of_agents))])

    lambd_k = rng.random((m,1))

    w_total_k = np.vstack([list_of_agents[i].x_k for i in range(len(list_of_agents))] + [lambd_k])
    w_total_k_prev = np.zeros((m*N + np.shape(lambd_k)[0], 1))

    norm_history = []

    norm = 0
    while (not norm_history or (norm:=np.linalg.norm(w_total_k - w_total_k_prev))>= EPSILON):
        norm_history.append(norm)
        
        if show_progress:
            print("norm =", norm)

            # intermediary plotting
            if (len(norm_history) in [2, 10, 50, 100, 150, 200]):
    #             list_of_agents[0].analyse_self()
                plot_agents(list_of_agents, save=True, n=len(norm_history))
        
        if len(norm_history)==1:
            print("plot for initial random agent decisions:")
            plot_agents(list_of_agents, save=False, n=len(norm_history))
        
        # individual agent iteration and updates
        for i in range(N):
            agent = list_of_agents[i]
            x_k_next = prox(agent, alpha_i, agent.x_k - alpha_i*(avg_x + c + (1/N)*agent.x_k + lambd_k))
            agent.d_k_next = 2*x_k_next - agent.x_k - b_i

            # update old agent's x_k
            agent.x_k = x_k_next

        # central coordinator
        avg_x = (1/N)*sum([list_of_agents[i].x_k for i in range(N)])

        lambd_k = np.maximum(ZEROS_VECTOR, lambd_k + beta*(1/N)*sum([list_of_agents[i].d_k_next for i in range(N)]))

        # update the x_total to recalculate the norm later
        w_total_k_prev = w_total_k
        w_total_k = np.vstack([list_of_agents[i].x_k for i in range(N)] + [lambd_k])
 
    # plot convergence of omega (w)
    # plt.plot(np.arange(0,len(norm_history)), norm_history)
    # plt.grid()

    # new figure, plot the energy usage on each time horizon
    # plt.figure(figsize=((6.4 * 1.6,4.8 * 1.6)))

    print("n iterations =", len(norm_history))
    return list_of_agents


if __name__ == "__main__":
    try:
        start = time.time()
        perform_pfb(show_progress=False)
        plot_agents(list_of_agents, save=False)
        print("time taken =", time.time() - start)
        
    except (KeyboardInterrupt):
        plot_agents(list_of_agents, save=False)
```
