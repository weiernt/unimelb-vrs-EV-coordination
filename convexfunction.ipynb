```python
%matplotlib notebook
import matplotlib.pyplot as plt
import numpy as np
def f(x):
    return ((x-3)**2)+2
if __name__ == "__main__":
    # plt.figure(figsize=((6.4 * 1.6,4.8 * 1.6)))
    plt.figure()
    plt.plot(np.linspace(0,10), f(np.linspace(0,10)))
    plt.title("Plot of $f(x)=(x-3)^2 + 2$")
    plt.ylabel("f(x)")
    plt.xlabel("x")
#     plt.grid()
#     plt.savefig("non_pev_demand.png")
```
